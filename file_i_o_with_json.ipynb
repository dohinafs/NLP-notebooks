```python
import json

with open("/content/claims_data.json") as f:
  data= json.load(f)

print(data)
```

```python
print(json.dumps(data,indent=2))
```

```python
pholder_name=[claim['policy_holder']['name'] for claim in data['claims']] #extract names of policy holders cnells
print(pholder_name)
```

```python
import pandas as pd
records=[]
for claim in data['claims']:
  record={
      'claim-id': claim['claim_id'],
      'name': claim['policy_holder']['name'],
      'email': claim['policy_holder']['contact']['email'],
      'loss-type': claim['loss_details']['type'],
      'loss-date': claim['loss_details']['date'],
      'status':claim['status']
  }
  records.append(record)
df=pd.DataFrame(records)
df
```

```python
for claim in data['claims']:
  if claim['claim_id']=="CL67890":
    claim['status']="APPROVED"

with open("/content/claims_new.json",'w') as f:
  json.dump(data,f,indent=2)
```
