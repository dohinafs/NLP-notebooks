```python
import pandas as pd
import numpy as np

data= pd.read_csv("/content/spam.csv",encoding='latin-1')[['v1','v2']]
text= data['v2']
label = data['v1']
print(data)
```

Tf-idf

```python
from sklearn.feature_extraction.text import TfidfVectorizer
corpus = list(text)
tfidf = TfidfVectorizer(max_features = 6000)
tfidf.fit(corpus)
tfidf_features = tfidf.transform(corpus)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import LinearSVC
from sklearn.metrics import precision_recall_fscore_support
# Converting the labels from strings to binary
le = LabelEncoder()
le.fit(label)
label = le.transform(label)
```

```python
train_percent = 0.7
train_cutoff = int(np.floor(train_percent*len(text) ) )
tfidf_model = LinearSVC()
tfidf_model.fit(tfidf_features[0 : train_cutoff],
                  label[0 : train_cutoff])
tfidf_prediction = tfidf_model.predict(
                  tfidf_features[train_cutoff + 1 : len(text)])
```

Results

```python
results = pd.DataFrame(index = ['TF-IDF'],
          columns = ['Precision', 'Recall', 'F1 score', 'support']
          )
results.loc['TF-IDF'] = precision_recall_fscore_support(
          label[train_cutoff + 1 : len(text)],
          tfidf_prediction,
          average = 'binary'
          )
print(results)
```

```python
test_text = ["Free money! Click here now!", "Hey, how are you doing today?"]

# Transform the test data using the same TF-IDF vectorizer
test_tfidf_features = tfidf.transform(test_text)

# Make predictions using the trained model
test_predictions = tfidf_model.predict(test_tfidf_features)

# Inverse transform the predictions to get the original labels back
predicted_labels = le.inverse_transform(test_predictions)

# View the original text and the predicted label
for i in range(len(test_text)):
    print(f"Text: {test_text[i]}")
    print(f"Predicted Label: {predicted_labels[i]}")
    print("-" * 20)
```
