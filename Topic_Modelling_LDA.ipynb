```python
import pandas as pd

df = pd.read_csv("data/recipes.csv")

#Replaces everything except English letters and spaces with a space — useful for cleaning text before tokenization or topic modeling.
df.content = df.ingredient_list.str.replace("[^A-Za-z ]", " ")

df.head()
```

```python
!pip install gensim==4.3.2 scipy==1.10.1 numpy==1.23.5 --force-reinstall --no-cache-dir
```

```python
from gensim.utils import simple_preprocess

texts = df.content.apply(simple_preprocess)
```

```python
from gensim import corpora

dictionary = corpora.Dictionary(texts)
dictionary.filter_extremes(no_below=5, no_above=0.5, keep_n=2000)
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
from gensim import models
```

```python
n=15
lda=models.LdaModel(corpus=corpus, num_topics=n)
```

```python
lda.print_topics()
```

```python
import pyLDAvis
import pyLDAvis.gensim
pyLDAvis.enable_notebook()
vis=pyLDAvis.gensim.prepare(lda, corpus, dictionary)
vis
```
