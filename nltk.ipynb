```python
AI="""Artificial Intelligence is transforming the way we interact with technology, enabling machines to perform tasks that once required human intelligence. From voice assistants
and recommendation systems to self-driving cars and medical diagnostics, AI is becoming an integral part of everyday life. By leveraging algorithms and vast amounts of data,
AI systems can recognize patterns, make decisions, and even improve over time through learning. As advancements continue, the line between human and machine capabilities is
becoming increasingly blurred, raising both exciting possibilities and important ethical considerations."""
```

```python
type(AI)
```

```python
from nltk.tokenize import word_tokenize
```

```python
Tokens = word_tokenize(AI)
Tokens
```

```python
len(Tokens)
```

Remove punctuation marks

```python
import re
punct = re.compile(r'[-.?!,:;()|0-9]')
```

```python
post_punct=[]
for i in Tokens:
  word=punct.sub("",i)
  if len(word)>0:
    post_punct.append(word)
post_punct
```

```python
len(post_punct)
```

Finding the frequency of each word

```python
from nltk.probability import FreqDist
dist=FreqDist()
```

```python
for i in Tokens:
  dist[i.lower()]+=1
dist
```

```python
for i in post_punct:
  dist[i.lower()]+=1
dist
```

```python
dist['intelligence']
```

```python
len(dist)
```

Top 10 most common

```python
top10= dist.most_common(10)
top10
```

```python
from nltk.util import bigrams, trigrams, ngrams
```

```python
string="The best and most beautiful things in the world cannot be seen or even touched, they must be felt with the heart."
q_tokens=word_tokenize(string)
q_tokens
```

```python
q_tokens=list(nltk.bigrams(q_tokens))
q_tokens
```

```python
q_tokens=word_tokenize(string)
q_tokens=list(nltk.trigrams(q_tokens))
q_tokens
```

```python
q_tokens=word_tokenize(string)
q_tokens=list(nltk.ngrams(q_tokens,5))
q_tokens
```

Stemming

```python
from nltk.stem import PorterStemmer
pst=PorterStemmer()
words=['giving', 'give', 'gave', 'given']
for i in words:
  print(i+ "-" +pst.stem(i))
```

```python
from nltk.stem import LancasterStemmer
lst=LancasterStemmer()
for i in words:
  print(i+"-"+lst.stem(i))
```

```python
from nltk.stem import SnowballStemmer
sbst=SnowballStemmer('english') #must include language
for i in words:
  print(i+"-"+sbst.stem(i))
```

```python
from nltk.stem import wordnet
from nltk.stem import WordNetLemmatizer
lem=WordNetLemmatizer()
words=['affection','affecting','affected']
for i in words:
  print(i+"-"+lem.lemmatize(i))
```

```python
lem.lemmatize('corpora')
```

POS Tag

```python
sent="He is a natural when it comes to painting."
s_tok=word_tokenize(sent)
s_tok
```

NER

```python
from nltk import ne_chunk
Nsent="The Queen stays in the Royal Palace of Madrid."
Ntok=word_tokenize(Nsent)
Ntok
```

```python
Npos=nltk.pos_tag(Ntok)
Npos
```

```python
ner_sent=ne_chunk(Npos)
print(ner_sent)
```

Chunking

```python
sent1="The big cat ate the little mouse which was after the cheese."
pos1=nltk.pos_tag(word_tokenize(sent1))
pos1
```

```python
grammar_np= r"NP: {<DT>?<JJ>*<NN>}"
```

```python
chunk_parse=nltk.RegexpParser(grammar_np)
chunk_parse
```

```python
result=chunk_parse.parse(pos1)
result
```
